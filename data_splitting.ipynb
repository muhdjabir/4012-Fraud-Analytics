```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

<h2> Reading in all the data

```python
orig_train = pd.read_csv("train_original.csv", sep="|")
orig_train.head()
```

```python
orig_test = pd.read_csv("test_original.csv", sep="|")
orig_test.head()
```

```python
labels = pd.read_csv("labels.csv")
labels.head()
```

<h2> Handling the data

```python
# Combining the labels with the test data
orig_test["fraud"] = labels["fraud"]
orig_test.head()
```

```python
all_data = pd.concat([orig_train, orig_test], ignore_index=True)
all_data
```

<h2> Splitting of the datasets

```python
train_dataset, test_dataset = train_test_split(all_data, test_size=0.3)
test_dataset, validate_dataset = train_test_split(test_dataset, test_size=1/3)
```
