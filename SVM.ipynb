#### Import Packages

```python
import pandas as pd
import numpy as np
import warnings
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score
from sklearn.inspection import permutation_importance
from scipy import stats
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 500)
```

#### Data Loading

```python
# Read CSV files into pandas DataFrames
test = pd.read_csv("C:/Users/Kaavya/OneDrive - National University of Singapore/Y3S1/BT4012/Final Project/test.csv")
train = pd.read_csv("C:/Users/Kaavya/OneDrive - National University of Singapore/Y3S1/BT4012/Final Project/train.csv")
validate = pd.read_csv("C:/Users/Kaavya/OneDrive - National University of Singapore/Y3S1/BT4012/Final Project/validate.csv")
```

```python
feature_cols = [col for col in train.columns if (col != 'fraud')]
target_col = ['fraud']
X_train = train[feature_cols].copy()
y_train = train[target_col].copy()
X_validate = validate[feature_cols].copy()
y_validate = validate[target_col].copy()
X_test  = test[feature_cols].copy()
y_test  = test[target_col].copy()
```

<h4> Feature Engineering

##### Total Number of Items Scanned

```python
train["totalItems"] = train["totalScanTimeInSeconds"] * train["scannedLineItemsPerSecond"]
test["totalItems"] = test["totalScanTimeInSeconds"] * test["scannedLineItemsPerSecond"]
validate["totalItems"] = validate["totalScanTimeInSeconds"] * validate["scannedLineItemsPerSecond"]
```

#### Feature Pre-processing

##### Scaling

```python
standardScaler = StandardScaler()
train[["standard_grandTotal", "standard_totalItems", "standard_totalScanTimeInSeconds", "standard_lineItemVoids", "standard_scansWithoutRegistration"]] = standardScaler.fit_transform(train[["grandTotal", "totalItems", "totalScanTimeInSeconds", "lineItemVoids", "scansWithoutRegistration"]])
test[["standard_grandTotal", "standard_totalItems", "standard_totalScanTimeInSeconds", "standard_lineItemVoids", "standard_scansWithoutRegistration"]] = standardScaler.transform(test[["grandTotal", "totalItems", "totalScanTimeInSeconds", "lineItemVoids", "scansWithoutRegistration"]])
validate[["standard_grandTotal", "standard_totalItems", "standard_totalScanTimeInSeconds", "standard_lineItemVoids", "standard_scansWithoutRegistration"]] = standardScaler.transform(validate[["grandTotal", "totalItems", "totalScanTimeInSeconds", "lineItemVoids", "scansWithoutRegistration"]])
```

<h5>Selected Features

```python
feature_cols = ["standard_grandTotal", "standard_totalItems", "standard_totalScanTimeInSeconds", "standard_lineItemVoids", "standard_scansWithoutRegistration",
                "trustLevel"]
target_col = ["fraud"]

X_train_selected_features = train[feature_cols].copy()
X_test_selected_features = test[feature_cols].copy()
X_validate_selected_features = validate[feature_cols].copy()

X_train_selected_features.head()
```

#### Linear Kernel - SVM

```python
linear_SVM = SVC(kernel="linear")

linear_SVM.fit(X_train_selected_features, y_train)
```

```python
train_predict = linear_SVM.predict(X_train_selected_features)
print("Train Data")
print('Accuracy score: {:0.4f}'.format(accuracy_score(y_train, train_predict)))
print('F1 score: {:0.4f}'.format(f1_score(y_train, train_predict)))
print('Precision score: {:0.4f}'.format(precision_score(y_train, train_predict)))
print('Recall score: {:0.4f}'.format(recall_score(y_train, train_predict)))
print("_________________________________________________________")

validation_predict = linear_SVM.predict(X_validate_selected_features)
print("Validate Data")
print('Accuracy score: {:0.4f}'.format(accuracy_score(y_validate, validation_predict)))
print('F1 score: {:0.4f}'.format(f1_score(y_validate, validation_predict)))
print('Precision score: {:0.4f}'.format(precision_score(y_validate, validation_predict)))
print('Recall score: {:0.4f}'.format(recall_score(y_validate, validation_predict)))
print("_________________________________________________________")
```

<h3>Hyperparameter Tuning

<h4> RandomSearch

```python
rand_list = {"C": stats.uniform(2, 10),
             "gamma": stats.uniform(0.1, 1)}
```

```python
# Set seed
np.random.seed(123)

random_model = RandomizedSearchCV(SVC(), rand_list, refit = True, n_iter = 10, n_jobs = 4, cv = 3, scoring='f1', verbose=3)

# fitting the model for random search
random_model.fit(X_train_selected_features, y_train)

print(random_model.best_estimator_)
```

```python
print(random_model.best_params_, random_model.get_params, random_model.decision_function)
```

```python
train_predict = random_model.predict(X_train_selected_features)
print("Train Data")
print('Accuracy score: {:0.4f}'.format(accuracy_score(y_train, train_predict)))
print('F1 score: {:0.4f}'.format(f1_score(y_train, train_predict)))
print('Precision score: {:0.4f}'.format(precision_score(y_train, train_predict)))
print('Recall score: {:0.4f}'.format(recall_score(y_train, train_predict)))
print("_________________________________________________________")

validation_predict = random_model.predict(X_validate_selected_features)
print("Validate Data")
print('Accuracy score: {:0.4f}'.format(accuracy_score(y_validate, validation_predict)))
print('F1 score: {:0.4f}'.format(f1_score(y_validate, validation_predict)))
print('Precision score: {:0.4f}'.format(precision_score(y_validate, validation_predict)))
print('Recall score: {:0.4f}'.format(recall_score(y_validate, validation_predict)))
print("_________________________________________________________")

test_predict = random_model.predict(X_test_selected_features)
print("Test Data")
print('Accuracy score: {:0.4f}'.format(accuracy_score(y_test, test_predict)))
print('F1 score: {:0.4f}'.format(f1_score(y_test, test_predict)))
print('Precision score: {:0.4f}'.format(precision_score(y_test, test_predict)))
print('Recall score: {:0.4f}'.format(recall_score(y_test, test_predict)))
print("_________________________________________________________")
```
